### Gen AI with OpenAI

```python
import os
from dotenv import load_dotenv

load_dotenv()

os.environ['OPENAI_API_KEY'] = os.getenv('OPENAI_API_KEY')
os.environ['LANGCHAIN_API_KEY'] = os.getenv('LANGCHAIN_API_KEY')
os.environ['LANGCHAIN_TRACING_V2'] = "true"
os.environ['LANGCHAIN_PROJECT'] = os.getenv('LANGCHAIN_PROJECT')

```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o")
print(llm)
```

```python
# Provide input and get the response

result = llm.invoke("What is Software engineering ?")
print(result.content)
```

### Lets use ChatPromptTemplate and StringParser


```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are an expert AI Engineer. Provide me answers based on the questions"),
        ("user", "{input}")
    ]
)
print(prompt)
```

```python
chain = prompt|llm
result = chain.invoke("How Bus operator makes money ?")
result.content
```

```python
from langchain_core.output_parsers import StrOutputParser

output_parser = StrOutputParser()
chain = prompt|llm|output_parser

result = chain.invoke({"input", "How Bus operator makes money ?"})
print(result)
```

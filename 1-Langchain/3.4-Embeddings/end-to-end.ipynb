### Lets do below 4 steps in Coding
1. Data Ingetion
2. Split text into Chunks
3. Apply Embedding Techniques(Open API, Ollama, HuggingFace etc.)
4. Store Ebeddings in VectorStore DB such as ChromaDB, FAISS etc.

```python
## Data Ingestion
from langchain_classic.document_loaders import TextLoader

loader = TextLoader("speech.txt")
docs = loader.load()
print(docs)
```

```python
## Text Splitter

from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter =  RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
split_document = text_splitter.split_documents(docs)

print(split_document)
```

```python
from langchain_openai import OpenAIEmbeddings

embedding_1024 = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
text = "This is a tutorial on OpenAI embedding"
query_result = embedding_1024.embed_query(text)
print(query_result)
```

```python
## Vector Embedding and Vector StoreDB

from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(split_document, embedding_1024)
print(db)

query = "Data ingestion is often the unsung hero of any AI pipeline"
retrieved_result = db.similarity_search(query)
print(retrieved_result)
```

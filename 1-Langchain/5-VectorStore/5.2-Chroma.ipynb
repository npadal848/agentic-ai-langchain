### Chroma
Chroma is an AI-native open-source vector database focused on developer productivity and happiness. Chroma is licensed under Apache 2.0.

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_ollama import OllamaEmbeddings
from langchain_chroma import Chroma

# Load data(speech.txt)
loader = TextLoader("speech.txt")
documents = loader.load()

# Split text into Chunks
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=30)
documents_splitter = text_splitter.split_documents(documents)

## Apply Ollama Embeddings to convert text chunks into vectors
embedding = OllamaEmbeddings(model="nomic-embed-text")

## Store chunks of vector in ChromaDB
chroma_db = Chroma.from_documents(documents_splitter, embedding)
print(chroma_db)
```

```python
query = "What is Langchain ?"
query_result = chroma_db.similarity_search(query)
print(query_result[0].page_content)
```

```python
## Save ChomaDB data into locally(disc)
chroma_db2 = Chroma.from_documents(documents_splitter, embedding, persist_directory="./chroma_db")
```

```python
# Load from Disk

db = Chroma(persist_directory="./chroma_db", embedding_function=embedding)
result = db.similarity_search(query)
print(result)
```
